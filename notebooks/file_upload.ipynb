```python
import boto3
import json

BUCKET = "jc-taekwondo-motion-cloud-computing-2026"
REGION = "us-east-2"

s3 = boto3.client("s3", region_name=REGION)

print("Files in S3 model/:")
resp = s3.list_objects_v2(Bucket=BUCKET, Prefix="model/")

for obj in resp.get("Contents", []):
    print(obj["Key"], obj["Size"], obj["LastModified"])

print("\nModel summary from S3:")
obj = s3.get_object(Bucket=BUCKET, Key="model/taekwondo_model_summary.json")
summary = json.loads(obj["Body"].read().decode("utf-8"))

print(json.dumps(summary, indent=2))
```
